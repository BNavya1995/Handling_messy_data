```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
df
```

```python
df['age']=df['age'].fillna(df['age'].median())
df['age']
```

```python
df['city']=df['city'].fillna(df['city'].mode()[0])
df['city']
```

```python
df=df.drop_duplicates()
df
```

```python
from sklearn.preprocessing import LabelEncoder
if 'city' in df.columns:
   df=pd.get_dummies(df,columns=['city'],prefix='city',dtype = int)
if 'gender' in df.columns:
  le=LabelEncoder()
  df['gender']=le.fit_transform(df['gender'])
print(df)
```

annual income=std =30903

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler_minmax=MinMaxScaler()
scaler_standard=StandardScaler()
num_col=['age','annual_income']

df_minmax=df.copy()
df_minmax[num_col]=scaler_minmax.fit_transform(df[num_col])
print("Min-Max Scaled DataFrame:\n ",df_minmax,"\n")

df_standard=df.copy()
df_standard[num_col]=scaler_standard.fit_transform(df[num_col])
print("Standarization Dataframe:\n",df_standard)
```
